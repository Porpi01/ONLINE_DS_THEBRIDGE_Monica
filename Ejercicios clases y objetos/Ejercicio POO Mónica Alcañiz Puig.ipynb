## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:

    #Atributos comunes
    Tipo = "Electrodomésticos"
    Abierta = True

    #Constructor con los atributos propios de cada tienda
    def __init__(self, nombre, direccion, numero_empleados, ventas_tres_meses):
        self.nombre = nombre
        self.direccion = direccion
        self.numero_empleados = numero_empleados
        self.venta_tres_meses = ventas_tres_meses

    #Método para calcular las ventas de todos los meses
    def calcular_ventas_totales(self):
        return sum(self.venta_tres_meses)

    #Método para calculaar la media de ventas de los últimos 3 meses
    def calcular_media_ventas_por_empleado(self):
         if self.numero_empleados > 0:
             return self.calcular_ventas_totales() / self.numero_empleados
         else:
             return 0

    #Método que devuelve en un string el nombre de la tienda, junto con su dirección.
    def obtener_info_tienda(self):
        return f"{self.nombre} - {self.direccion}"

    #Método que obtiene las ventas del último mes.
    def obtener_ventas_ultimo_mes(self):
        if self.venta_tres_meses:
            return self.venta_tres_meses[-1]
        else:
            return 0

    #Método para dar una proyección de las ventas en caso de haber invertido
    def proyectar_ventas_con_marketing(self, inversion_marketing):
        ventas_totales = self.calcular_ventas_totales()
        if inversion_marketing < 1000:
            factor_aumento = 1.2
        else:
            factor_aumento = 1.5
        nuevas_ventas = [venta * factor_aumento for venta in self.venta_tres_meses]
        self.venta_tres_meses = nuevas_ventas
        return sum(nuevas_ventas)

# Crear tres tiendas con datos inventados
tienda1 = Tienda("CompramosTuCoche", "Calle Juan Ramón Giner 15", 3, [15000, 18000, 16500])
tienda2 = Tienda("PencilBase", "Avenida de la Estacioneta 17", 5, [22000, 25000, 23000])
tienda3 = Tienda("MundoDigit", "Plaza del Sol 5", 6, [9000, 10500, 9800])

# Comprobar todo lo implementado en la clase Tienda para tienda1
print("Nombre:", tienda1.nombre)
print("Dirección:", tienda1.direccion)
print("Número de empleados:", tienda1.numero_empleados)
print("Ventas últimos tres meses:", tienda1.venta_tres_meses) 
print("Ventas totales:", tienda1.calcular_ventas_totales())
print("Media de ventas por empleado:", tienda1.calcular_media_ventas_por_empleado())
print("Información de la tienda", tienda1.obtener_info_tienda())
print("Ventas del último mes", tienda1.obtener_ventas_ultimo_mes())

inversion = 1000
proyeccion = tienda1.proyectar_ventas_con_marketing(inversion)
print("Proyección de ventas con inversión de", inversion, ":", proyeccion)
print("Nuevas ventas últimos tres meses", tienda1.venta_tres_meses)

# Calcular las ventas del último mes de todas las tiendas
print("Ventas del último mes de todas las tiendas")
tiendas = [tienda1, tienda2, tienda3]
for tienda in tiendas:
    print("Ventas del último mes de", tienda.nombre, ":", tienda.obtener_ventas_ultimo_mes())

# Imprimir los nombres de las tiendas cuya dirección lleve el string "Avenida"
print("Tiendas en Avenida")
for tienda in tiendas:
    if "Avenida" in tienda.direccion:
        print(tienda.nombre)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    
 #Atributos
    patas = 4
    orejas = 2
    ojos = 2
    
#Constructor
    def __init__(self, raza, pelo="Marrón", dueño=None):
        self.raza = raza
        self.pelo = pelo
        self.dueño = dueño
        self.velocidad = 0
#Método andar para incrementar la velocidad del perro
    def andar(self, aumento_velocidad):
     
        self.velocidad += aumento_velocidad
#Método parar establece la velocidad del perro en 0
    def parar(self):
        self.velocidad = 0
#Método ladrar devuelve el ladrido del perro junto con un sonido extra
    def ladrar(self, sonido_extra):
        return "GUAU!" + sonido_extra

# Crear un objeto de tipo Perro, sin dueño
mi_perro = Perro("Beagle")

# Comprobar que están bien todos sus atributos

print("Raza:", mi_perro.raza)
print("Pelo:", mi_perro.pelo)
print("Dueño:", mi_perro.dueño)
print("Patas:", mi_perro.patas)
print("Orejas:", mi_perro.orejas)
print("Ojos:", mi_perro.ojos)
print("Velocidad inicial:", mi_perro.velocidad)

# Prueba que ande, y comprueba su velocidad
mi_perro.andar(5)
print("Velocidad después de andar:", mi_perro.velocidad)

# Páralo
mi_perro.parar()
print("Velocidad después de parar:", mi_perro.velocidad)

# Prueba el método ladrar
sonido = mi_perro.ladrar(" GRRR!")
print("Sonido del perro:", sonido)
```
