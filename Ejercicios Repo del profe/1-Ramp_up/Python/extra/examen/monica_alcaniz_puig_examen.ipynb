```python

# 1.Calculo de distancia con función

#Función calcular distancia con dos parámetros tiempo y gravedad para calcular la altura_inicial
def calcular_distancia_caida_libre (tiempo_caida, gravedad = 9.8):
    distancia = (1/2)*gravedad*tiempo_caida**2
    return distancia


tiempo_caida = float(input("Introduce el tiempo"))
calcular_distancia_caida_libre(tiempo_caida)  
```

```python
import math

#Función calcular tiempo con dos parámetros altura y gravedad 

def calcular_tiempo_caida_libre(altura_inicial, gravedad = 9.8):
    tiempo = math.sqrt((2*altura_inicial)/gravedad)
    return tiempo

altura_inicial = float(input("Introduce la altura"))

calcular_tiempo_caida_libre(altura_inicial)
```

```python
import math

#Función calcular velocidad con dos parámetros altura y gravedad 

def calcular_velocidad_final (altura_inicial, gravedad = 9.8):
    velocidad_final = math.sqrt(2*gravedad*altura_inicial)
    return velocidad_final

altura_inicial = float(input("Introduce la altura inicial"))
calcular_velocidad_final(altura_inicial)
```

## Ejercicio 4: Simulación de un videojuego con clases en Python (2 puntos)

En este ejercicio, se creará una clase para simular un personaje en un videojuego.

**1. Diseño de la clase:**

- **Nombre:** `Personaje`

**Atributos:**

* `nombre`: Nombre del personaje (str)
* `vida`: Puntos de vida (int) 50 o 100 por ejemplo
* `nivel`: Nivel del personaje (int) 1
* `experiencia`: Experiencia acumulada (int) 0

**Métodos:**

* `mision(dificultad)`: Gana 10, 20 o 30 de experiencia (llamar metodo ganar experiencia) en funcion de la dificultad ("baja", "media", "alta").
* `defender(ataque)`: Reduce el daño recibido del enemigo en función del nivel del personaje (vida -= ataque/nivel)
* `ganar_experiencia(experiencia)`: Aumenta la experiencia del personaje. Si la experiencia acumulada supera un umbral (50), el personaje sube de nivel (llama a subir_nivel).
* `subir_nivel()`: Aumenta el nivel del personaje +1 y sus puntos de vida un 10%. Resetea la experiencia a 0.







```python
# CREA UN PERSONAJE
# QUE HAGA UN PAR DE MISIONES
# QUE SE DEFIENDA UN PAR DE VECES 
# HASTA QUE AL MENOS SUBA DE NIVEL UNA VEZ

class Personaje:
   
   def __init__(self,nombre,vida = 100, nivel = 1, experiencia = 0):
       self.nombre = nombre
       self.vida = vida
       self.nivel = nivel
       self.experiencia = experiencia
    

   def mision(self, dificultad):
       if dificultad == "baja":
         self.ganar_experiencia(10)
       elif dificultad == "media":
         self.ganar_experiencia(20)
       else:
         self.ganar_experiencia(30)
      

   def defender(self,ataque):
       daño = ataque/self.nivel
       self.vida -= daño
       print(f"Tu daño ha sido de {daño}")
       return daño

   def ganar_experiencia(self,experiencia):
       self.experiencia += experiencia
       print(f"Tu nivel de experiencia actual es {self.experiencia}")

       if self.experiencia > 50:
          self.subir_nivel()
       else:
          print("No tienes la habilidad suficiente para subir de nivel")
  

   def subir_nivel(self):
       nivel += 1
       vida = vida + (0,1*vida)
       print(f"Tu nivel actual es {nivel} y tu vida actual es {vida}")
       return nivel,vida
  

personaje = Personaje("Pablo Motos")
personaje.mision("baja")
```
