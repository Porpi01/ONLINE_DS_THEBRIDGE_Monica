# Occupation

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
mask = users.loc[:,"gender"] == "M"

male_count = users[mask].groupby("occupation").count()
total = users.groupby("occupation").count()

male_ratio = male_count/total
male_ratio.sort_values(by="gender", ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].min()
users.groupby("occupation")["age"].max()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])["age"].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
mask1 = users.loc[:,"gender"] == "F"
mask2 = users.loc[:,"gender"] == "M"

women = users[mask1].groupby("occupation")["gender"]
male = users[mask2].groupby("occupation")["gender"]


total_count = users.groupby("occupation")["gender"].count()


percentage_women = women/total_count
percentage_male = male/total_count
```
