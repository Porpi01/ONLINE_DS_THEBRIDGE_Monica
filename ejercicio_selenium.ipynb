```python
# Mostrar en un dataframe 250 peliculas. Año nombre, director, actores y nota

from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

```python
driver = webdriver.Chrome()

url = 'https://www.filmaffinity.com/es/topgen.php'

driver.get(url)
```

```python

botoncookies = driver.find_element(By.XPATH, '//*[@id="accept-btn"]')
botoncookies.click()
```

```python
for _ in range(9):
    boton_scroll = driver.find_element(By.XPATH, '//*[@id="load-more-bt"]')
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    boton_scroll.click()
    time.sleep(1)
```

```python
sopa = bs(driver.page_source)

lista_titulos = []
lista_años = []
lista_directores = []
lista_actores = []
lista_notas = []

titulos = sopa.select('div.mc-title')
for t in titulos:
    titulo = t.find('a').get_text()
    lista_titulos.append(titulo)

años_publicacion = sopa.select('span.mc-year')
for año in años_publicacion:
    años = año.get_text()
    lista_años.append(años)

directores = sopa.select('div.mc-director')
for dic in directores:
    director = dic.find('a').get_text()
    lista_directores.append(director)

actores = sopa.select('div.mc-cast')
for act in actores:
    elenco = act.text.strip().replace("\n", "").replace("  ", "")
    lista_actores.append(elenco)

score = sopa.select('div.avg-rating')
for nota in score:
    notas = nota.get_text().replace(",",".")
    lista_notas.append(notas)
datos = { "titulo": lista_titulos, "fecha_de_estreno": lista_años, "director": lista_directores, "elenco": lista_actores, "puntuacion": lista_notas

}

df = pd.DataFrame(datos)
df
```
